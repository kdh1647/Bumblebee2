```python
import pandas as pd
import json
import requests
```

```python
url = 'http://ecos.bok.or.kr/api/'
#url = 'https://ecos.bok.or.kr/api/'  DRB 네트워크에서 SSL인증문제로 https 안됨
```

```python
#Method Option 
ServiceName = 'StatisticSearch'    #API서비스명
Key = 'TZ39XYZ04A7GUAGUPSMF'            #한국은행에서 발급받은 오픈API 인증키
FileType = 'json'   #결과값의 파일 형식 - xml, json
Lan = 'kr'          #결과값의 언어 - kr(국문), en(영문)
ReqStart = '1'      #전체 결과값 중 시작 번호
ReqEnd = '100'       #전체 결과값 중 끝 번호
StaticCode = '513Y001'    #통계표코드 참고 https://ecos.bok.or.kr/api/#/DevGuide/StatisticalCodeSearch 
Period = 'M'              #주기(년:A, 반년:S, 분기:Q, 월:M, 반월:SM, 일: D)
StartYear = '201701'      #검색시작일 
EndYear = '202210'        #검색종료
ItemCode1 = 'E1000'       #통계항목코드
ItemCode2 = '?'
ItemCode3 = '?'
ItemCode4 = '?'

#https://ecos.bok.or.kr/api/StatisticSearch/sample/xml/kr/1/10/200Y001/A/2015/2021/10101/?/?/?
```

```python
ApiUrl = ''
ApiUrl = url + ServiceName + '/' + Key + '/' + FileType + '/' + Lan + '/' + ReqStart + '/' + ReqEnd + '/' + StaticCode + '/'  
ApiUrl = ApiUrl + Period + '/' + StartYear + '/' + EndYear + '/' + ItemCode1 + '/' + ItemCode3 + '/' + ItemCode4 

ApiUrl
```

```python
result = requests.get(ApiUrl)
```

```python
js = json.loads(result.content)
js
```

```python
tmp = pd.DataFrame(js['StatisticSearch']['row'])    #json -> DataFrame
```
